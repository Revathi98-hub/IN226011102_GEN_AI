## **2. Load Free LLM**

```python
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

# Load lightweight free model
pipe = pipeline(
    "text-generation",
    model="mistralai/Mistral-7B-Instruct-v0.1",
    max_new_tokens=100
)

llm = HuggingFacePipeline(pipeline=pipe)

print(llm.invoke("Explain LangChain in one line"))
```

# **Introduction to LangChain**

LangChain is a framework that helps developers build intelligent applications using Large Language Models (LLMs) by connecting multiple components like prompts, memory, tools, and external data.

Instead of treating LLMs as standalone text generators, LangChain enables structured workflows.

## **Why It Matters**

Real-world AI applications require:
- Multi-step reasoning
- Context awareness
- External data integration

LangChain solves:
- Orchestration of workflows
- Chaining operations
- Tool integration
- Memory handling

In simple terms:
LLM = Brain  
LangChain = System that makes the brain useful

## **Prompt Templates**

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template(
    "Explain {topic} in simple terms"
)

print(prompt.format(topic="LangChain"))
```

## **Chains**

```python
from langchain_core.output_parsers import StrOutputParser

chain = prompt | llm | StrOutputParser()

print(chain.invoke({"topic": "LangChain Chains"}))
```

## **Tools**

```python
from langchain.tools import tool

@tool
def calculator(expr: str) -> str:
    """Evaluates a mathematical expression and returns the result."""
    return str(eval(expr))

print(calculator.run("10 * 5"))
```

# **Real-World Use Cases**

## **Use Case 1: Customer Support Chatbot with Memory**
**Problem Statement**

In traditional customer support systems, maintaining context across conversations is a major challenge. Users often have to repeat their information such as their name, previous issues, or order details multiple times because the system does not retain prior interactions. This leads to frustration, increased resolution time, and a poor user experience, especially in scenarios where multiple queries are connected.

A stateless chatbot treats every message as a new interaction, which makes it incapable of handling real-world conversational flows. In customer support environments where continuity and personalization are critical, this limitation significantly reduces efficiency and user satisfaction.

**Why a Plain LLM is Not Enough**

A basic Large Language Model processes each input independently without remembering past interactions unless explicitly provided. This means it cannot maintain conversation history or track user-specific details across multiple turns. As a result, even if the responses are accurate, they lack continuity and personalization.

Without memory, the chatbot behaves like it is interacting with a new user every time, which breaks the natural flow of conversation. This makes it unsuitable for applications like customer support, where understanding context is essential for resolving issues effectively.

**Solution Using LangChain**

LangChain introduces memory components that allow the chatbot to store and reuse previous interactions. By maintaining a conversation buffer, the system can include past messages as part of the current input, enabling the model to generate context-aware responses.

This approach transforms the chatbot into a more intelligent system that can remember user details, maintain conversation flow, and provide personalized assistance. It significantly improves the quality of interaction and aligns the chatbot behavior with real-world customer support expectations.

## **Use Case 2: Document Q&A System (RAG)**
 **Problem Statement**

Large Language Models are trained on pre-existing data and do not have access to real-time or domain-specific information such as company documents, research papers, or internal knowledge bases. This creates a major limitation when users ask questions related to specific or updated content.

In such scenarios, relying solely on the model can lead to incorrect or fabricated answers, commonly known as hallucinations. This makes plain LLMs unreliable for applications that require factual accuracy and domain relevance.

**Why a Plain LLM is Not Enough**

A standalone LLM does not have access to external documents unless the information is explicitly included in the prompt. This restricts its ability to answer questions based on private or newly updated data, which is often required in real-world applications.

Additionally, without retrieval mechanisms, the model may generate plausible but incorrect answers. This lack of grounding in actual data reduces trust and makes it unsuitable for knowledge-intensive tasks.

**Solution Using LangChain (RAG)**

LangChain enables Retrieval-Augmented Generation (RAG), where external documents are first converted into embeddings and stored in a vector database. When a query is received, the system retrieves the most relevant information and provides it to the model as context.

This approach ensures that responses are grounded in actual data rather than relying solely on the model’s internal knowledge. It improves accuracy, reduces hallucinations, and allows the system to work with domain-specific or private datasets.

## **Use Case 3: AI Personal Assistant**
**Problem Statement**

Modern users expect AI systems to go beyond answering questions and actively assist in performing tasks such as scheduling, calculations, and information retrieval. However, traditional chatbots are limited to text generation and cannot execute real-world actions.

This limitation prevents them from functioning as true assistants, as they lack the ability to interact with external tools or perform multi-step reasoning required for complex tasks.

**Why a Plain LLM is Not Enough**

A basic LLM can generate responses but cannot perform actions like calculations, API calls, or workflow execution. It lacks the ability to make decisions or interact with external systems, which limits its practical usability.

Without these capabilities, the assistant remains passive and cannot handle tasks that require execution or dynamic problem-solving, making it less effective for real-world applications.

**Solution Using LangChain**

LangChain enables the integration of tools and decision-making mechanisms, allowing the system to act based on user input. By combining LLMs with tools and structured workflows, the assistant can perform tasks such as calculations, data retrieval, and automation.

This transforms the assistant into an active system capable of both understanding queries and executing actions. It enables multi-step reasoning and dynamic task handling, which are essential for building intelligent assistants.
